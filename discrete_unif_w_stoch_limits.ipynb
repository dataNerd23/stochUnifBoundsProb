# Import Packages

```python
from scipy.stats import randint
import plotly.graph_objects as go
```

# Initial Set up

```python
L1 = 1
U1 = 20
U2 = 40
```

# Exact Distribution
Calculating the exact distribution of $X \sim \text{discrete }U(y,z)$ where $Y \sim \text{discrete }U(L_1,U_1)$ and $Z \sim \text{discrete }U(U_1+1,U_2)$ where $L_1 < U_1 < U_2$:

$p(X=x)=\left(\frac{1}{U_1-L_1+1}\right)\left(\frac{1}{U_2-U_1}\right)\sum\limits_{y=\min(x,U_1+1)}^{U_2}\sum\limits_{z=L_1}^{\max(U_1,x)}\frac{1}{y-z+1}$

```python
f_x = []
denom = ((U1-L1+1)*(U2-U1))
for x in range(L1,U2+1):
    value = 0
    for y in range(max(x,U1+1),U2+1):
        for z in range(1,min(x+1,U1+1)):
            value += 1/(y-z+1)
    f_x.append(value/denom)

trace = go.Bar(x=list(range(L1,U2+1)),
               y=f_x)
fig = go.Figure(trace)
fig.show()
```

# Simulatating Distribution

```python
yList = randint.rvs(U1+1, U2+1, size=100)
zList = randint.rvs(L1, U1+1, size=100)
xList = []

for yRand in yList:
    for zRand in zList:
        xList.append(randint.rvs(zRand, yRand+1, size=1)[0])

fig = go.Figure(data=[go.Histogram(x=xList,histnorm='probability')])
fig.show()
```
